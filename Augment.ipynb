```python
import xml.etree.ElementTree as ET
import os
import cv2
import numpy as np

#Parameter

main_path = "D:/Sekolah/Skripsi/Dataset/Augment/"
```

```python

#Augment Data Flip Horizontal
num_data = int(len(os.listdir(main_path))/2)
directory = sorted(os.listdir(main_path))
for ldir in directory:
    if "xml" in ldir:
        xml_path = main_path + ldir
        tree = ET.parse(xml_path)
        for elem in tree.iter():
            if 'filename' in elem.tag:
                print(elem.text)
                index = int(''.join([n for n in elem.text if n.isdigit()]))
                new_index = num_data + index
                new_index = str(new_index)
                if "spanner" in elem.text:
                    file_name = "spanner_" + new_index + ".jpg"
                    xml_name = "spanner_" + new_index + ".xml"
                    print("New filename : {}".format(file_name))
                    tree.find('.//filename').text = file_name
                elif "cutter" in elem.text:
                    file_name = "cutter_" + new_index + ".jpg"
                    xml_name = "cutter_" + new_index + ".xml"
                    print("New filename : {}".format(file_name))
                    tree.find('.//filename').text = file_name
                elif "palu" in elem.text:
                    file_name = "palu_" + new_index + ".jpg"
                    xml_name = "palu_" + new_index + ".xml"
                    print("New filename : {}".format(file_name))
                    tree.find('.//filename').text = file_name
                elif "obeng_solder" in elem.text:
                    file_name = "obeng_solder_" + new_index + ".jpg"
                    xml_name = "obeng_solder_" + new_index + ".xml"
                    print("New filename : {}".format(file_name))
                    tree.find('.//filename').text = file_name
                
            if 'object' in elem.tag:
                for attr in list(elem):
                    if 'bndbox' in attr.tag:
                        for dim in list(attr):
                            if 'xmin' in dim.tag:
                                x1 = int(round(float(dim.text)))
                                x1_ = x1
                            if 'ymin' in dim.tag:
                                y1 = int(round(float(dim.text)))
                            if 'xmax' in dim.tag:
                                x2 = int(round(float(dim.text)))
                                x2_ = x2
                            if 'ymax' in dim.tag:
                                y2 = int(round(float(dim.text)))
                
                x2 = 300 - x1_
                x1 = 300 - x2_                            
                print(x1_, y1, x2_, y2)
                tree.find('.//xmin').text = str(x1)
                tree.find('.//ymin').text = str(y1)
                tree.find('.//xmax').text = str(x2)
                tree.find('.//ymax').text = str(y2)

        tree.write(main_path+xml_name)

    else:
        img_path = main_path + ldir

        index = int(''.join([n for n in ldir if n.isdigit()]))
        new_index = num_data + index
        new_index = str(new_index)
        if "spanner" in ldir:
            img_name = "spanner_" + new_index + ".jpg"
            print("New img name : {}".format(img_name))
        elif "cutter" in ldir:
            img_name = "cutter_" + new_index + ".jpg"
            print("New img name : {}".format(img_name))
        elif "palu" in ldir:
            img_name = "palu_" + new_index + ".jpg"
            print("New img name : {}".format(img_name))
        elif "obeng_solder" in ldir:
            img_name = "obeng_solder_" + new_index + ".jpg"
            print("New img name : {}".format(img_name))

        img = cv2.imread(img_path)
        img = cv2.flip(img, 1)
        cv2.imwrite(main_path + img_name, img)
```

```python

#Augment Data Kompleks
num_data = int(len(os.listdir(main_path))/2)
directory = sorted(os.listdir(main_path))
i=0
for ldir in directory:
    if "xml" in ldir:
        xml_path = main_path + ldir
        tree = ET.parse(xml_path)
            
        for elem in tree.iter():
            if 'filename' in elem.tag:
                
                index = int(''.join([n for n in elem.text if n.isdigit()]))
                new_index = num_data + index
                new_index = str(new_index)
                if "spanner" in elem.text:
                    file_name = "spanner_" + new_index + ".jpg"
                    xml_name = "spanner_" + new_index + ".xml"
                    print("New filename : {}".format(file_name))
                    tree.find('.//filename').text = file_name
                elif "obeng_solder" in elem.text:
                    file_name = "obeng_solder_" + new_index + ".jpg"
                    xml_name = "obeng_solder_" + new_index + ".xml"
                    print("New filename : {}".format(file_name))
                    tree.find('.//filename').text = file_name
                
            if 'object' in elem.tag:
                for attr in list(elem):
                    if 'name' in attr.tag:
                        attr_name = attr.text
                        print(attr_name)
                    if 'bndbox' in attr.tag:
                        for dim in list(attr):
                            if 'xmin' in dim.tag:
                                x1 = int(round(float(dim.text)))
                                x1_ = x1
                            if 'ymin' in dim.tag:
                                y1 = int(round(float(dim.text)))
                            if 'xmax' in dim.tag:
                                x2 = int(round(float(dim.text)))
                                x2_ = x2
                            if 'ymax' in dim.tag:
                                y2 = int(round(float(dim.text)))
                
                        x2 = 300 - x1_
                        x1 = 300 - x2_ 
                        print(x1_, y1, x2_, y2)
                        print(x1, y1, x2, y2) 
                        for obj in tree.findall('object'):
                            if obj.find('name').text == "solder" and attr_name == "solder":
                                for bnd in obj.findall('bndbox'):
                                    bnd.find('xmin').text = str(x1)
                                    bnd.find('ymin').text = str(y1)
                                    bnd.find('xmax').text = str(x2)
                                    bnd.find('ymax').text = str(y2)
                            elif obj.find('name').text == "obeng" and attr_name == "obeng" :
                                for bnd in obj.findall('bndbox'):
                                    bnd.find('xmin').text = str(x1)
                                    bnd.find('ymin').text = str(y1)
                                    bnd.find('xmax').text = str(x2)
                                    bnd.find('ymax').text = str(y2)
        tree.write(main_path+xml_name)

    else:
        img_path = main_path + ldir

        index = int(''.join([n for n in ldir if n.isdigit()]))
        new_index = num_data + index
        new_index = str(new_index)
        if "obeng_solder" in ldir:
            img_name = "obeng_solder_" + new_index + ".jpg"
            print("New img name : {}".format(img_name))

        img = cv2.imread(img_path)
        img = cv2.flip(img, 1)
        cv2.imwrite(main_path + img_name, img)
```
